Напишіть програму, яка приймає на вхід число і в залежності від умов виводить текст YES або NO.

Умови:

**якщо число непарне, вивести «YES»;**

**якщо число парне в діапазоні від 2 до 5 (включно), вивести «NO»;**

**якщо число парне в діапазоні від 6 до 20 (включно), вивести «YES»;**

**якщо число парне і більше 20, вивести «NO».**

**Формат вхідних даних**

На вхід до програми подається натуральне число.

**Формат вихідних даних**

Програма має вивести текст відповідно до умови завдання.

**Sample Input 1:**

1

**Sample Output 1:**

YES

**Sample Input 2:**

2

**Sample Output 2:**

NO

```python
for _ in range(4):
    number = int(input("Enter a number: "))

    if number % 2 == 1:
        print("YES")
    elif 2 <= number <= 5:
        print("NO")
    elif 6 <= number <= 20:
        print("YES")
    elif number > 20 and number % 2 == 0:
        print("NO")
```

**Хід ферзя**

Дано дві різні клітини шахової дошки. Напишіть програму, яка визначає, чи може ферзь потрапити з першої клітини на другу одним ходом. Програма отримує на вхід чотири числа від 1 до 8 кожне, що задають номер стовпця і рядка спочатку для першої клітини, потім для другої клітини. Програма повинна вивести YES, якщо з першої клітини ходом ферзя можна потрапити в другу або NO в іншому випадку.

**Формат вхідних даних**

На вхід до програми подається чотири числа від 1 до 8.

**Формат вихідних даних**

Програма має вивести текст відповідно до умови завдання.

**Примітка**. Шаховий ферзь ходить по діагоналі, горизонталі чи вертикалі.

Playing field:

```python
start = input("Enter the initial location of the shape:")
x1 = ord(start[0]) - ord('a') + 1
y1 = int(start[1])

end = input("Enter the target location of the shape:")
x2 = ord(end[0]) - ord('a') + 1
y2 = int(end[1])

# Checking whether the queen can get from the first square to the second in one move
if x1 == x2 or y1 == y2 or abs(x1 - x2) == abs(y1 - y2):
    print("YES")
else:
    print("NO")
```
